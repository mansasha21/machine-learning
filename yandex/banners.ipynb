```python
import pandas as pd
from sklearn.cluster import mean_shift
import numpy as np
```

```python
df = pd.read_csv('checkins.dat',sep='|')
df.head()
```

```python
df=df.dropna()
```

```python
df.set_axis(['id', 'user_id', 'venue_id', 'latitude',
       'longitude', 'created_at'],axis='columns', inplace=True)
```

```python
resdf = pd.DataFrame(df.columns)
```

```python
df.latitude = df.latitude.apply(lambda x : None if len(x.strip())==0 else float(x))
df.longitude = df.longitude.apply(lambda x : None if len(x.strip())==0 else float(x))
```

```python
df.dropna(inplace=True)
x = df[['latitude','longitude']]
```

```python
%time km = mean_shift(x[:100000].values, bandwidth=0.1)
```

```python
cities = [[33.751277, -118.188740], [25.867736, -80.324116], [51.503016, -0.075479], [52.378894, 4.885084], [39.366487, 117.036146], [-33.868457, 151.205134]]
```

```python
%time dis = [[((pos1[0]-pos2[0])**2 + (pos1[1] - pos2[1])**2)**(0.5), pos1] for pos1 in km[0] for pos2 in cities]
```
