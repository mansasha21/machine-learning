```python
from sklearn import datasets, ensemble, metrics, model_selection,tree
from sklearn.datasets import load_digits
import numpy as np
import matplotlib.pyplot as plt
```

```python
digits = load_digits()
data = digits['data']
target = digits['target']
data.shape
#model_selection.cross_val_score(cv=10) mean
```

```python
classifier = tree.DecisionTreeClassifier()
score = model_selection.cross_val_score(classifier,data,target, cv=10)
score.mean()
```

```python
fitted_classifier = ensemble.BaggingClassifier(classifier,n_estimators=100)
```

```python
score = model_selection.cross_val_score(fitted_classifier,data,target, cv=10)
score.mean()
```

```python
classifier = tree.DecisionTreeClassifier()
fitted_classifier_finite = ensemble.BaggingClassifier(classifier,n_estimators=100,max_features=8)
score = model_selection.cross_val_score(fitted_classifier_finite,data,target, cv=10)
score.mean()
```

```python
classifier = tree.DecisionTreeClassifier(max_features=8)
fitted_classifier_finite = ensemble.BaggingClassifier(classifier,n_estimators=100)
score = model_selection.cross_val_score(fitted_classifier_finite,data,target, cv=10)
score.mean()
```

```python
forests = [ensemble.RandomForestClassifier(n_estimators = i*10, max_depth = 10, random_state = 1) for i in range(1,10)]
for f in forests:
    f.fit(data,target)
score =  [model_selection.cross_val_score(f,data,target, cv=10).mean() for f in forests]
plt.plot(range(10,100,10),score)
plt.show()
    
```

```python
forests = [ensemble.RandomForestClassifier(n_estimators = 60, max_depth = 5,max_features=i*6, random_state = 1) for i in range(1,10)]
for f in forests:
    f.fit(data,target)
score =  [model_selection.cross_val_score(f,data,target, cv=10).mean() for f in forests]
plt.plot(range(10,100,10),score)
plt.show()
```
