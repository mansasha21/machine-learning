```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import gamma, norm
%matplotlib inline
```

Сгенерируем выборку объёма 1000 из гамма распределения с параметром 2 и построим гистограмму и плотность распределения.
https://ru.wikipedia.org/wiki/Гамма-распределение

```python
fig, ax = plt.subplots(1, 1)
param = 2
sample = gamma.rvs(param,size =1000)
x = np.linspace(0,10,1000)
pdf = gamma.pdf(param,x)
ax.plot(x, pdf, label='theoretical pdf')
ax.hist(sample,density=True)
ax.legend(loc='best', frameon=False)
plt.show()
```

Сгенировал 1000 выборок объема 5/25/100 и построил гистограммы распределения их средних выборочных значений.

```python
sample =[gamma.rvs(param,size = 5) for i in range(1000)]
sample_mean = [np.mean(i) for i in sample]
sample1 =[gamma.rvs(param,size = 25) for i in range(1000)]
sample_mean1 = [np.mean(i) for i in sample1]
sample2 =[gamma.rvs(param,size = 100) for i in range(1000)]
sample_mean2 = [np.mean(i) for i in sample2]
```

EX = 2, DX = sqrt(2/N) где N объем выборки, 2 - параметр гамма распределения 

```python
fig, ax = plt.subplots(1, 1)

norm_rv = sts.norm(2, 0.2**(1/2))
x = np.linspace(0,10,5)
pdf = norm_rv.pdf(x)
ax.plot(x, pdf, label='pdf')
ax.hist(sample_mean,density=True)
ax.legend(loc='best', frameon=False)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
norm_rv = sts.norm(2, 0.08**(1/2))
x = np.linspace(0,10,25)
pdf = norm_rv.pdf(x)
ax.plot(x, pdf, label='pdf')
ax.hist(sample_mean1,density=True)
ax.legend(loc='best', frameon=False)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
norm_rv = sts.norm(2, 0.02**(1/2))
x = np.linspace(0,10,100)
pdf = norm_rv.pdf(x)
ax.plot(x, pdf, label='pdf')
ax.hist(sample_mean2,density=True)
ax.legend(loc='best', frameon=False)
plt.show()
```
